```python
import os
from pathlib import Path
from icrawler.builtin import BingImageCrawler

icloud_base = Path.home() / "Library/Mobile Documents/com~apple~CloudDocs"
base_dir = icloud_base / "Desktop/Python_AI and Unreal Workshop/Web Scraping"
pic_dir = base_dir / "Pictures"
film_dir = base_dir / "Films"

os.makedirs(pic_dir, exist_ok=True)
os.makedirs(film_dir, exist_ok=True)

def gothic_bing_crawler():
    print("Downloading Gothic architecture images...")
    arch_crawler = BingImageCrawler(storage={'root_dir': str(pic_dir)})
    arch_crawler.crawl(keyword="Gothic cathedral exterior architecture", max_num=300)

    print("\nDownloading Gothic film stills...")
    film_crawler = BingImageCrawler(storage={'root_dir': str(film_dir)})
    film_crawler.crawl(keyword="Gothic horror movie cinematography stills", max_num=200)

    print("\nDownload complete. Check folders.")

gothic_bing_crawler()

---
import os
from pathlib import Path
from icrawler.builtin import GoogleImageCrawler

icloud_base = Path.home() / "Library/Mobile Documents/com~apple~CloudDocs"
base_dir = icloud_base / "Desktop/Python_AI and Unreal Workshop/Web Scraping"

pic_dir = base_dir / "Pictures"
film_dir = base_dir / "Films"

os.makedirs(pic_dir, exist_ok=True)
os.makedirs(film_dir, exist_ok=True)

def gothic_dual_crawler():
    print(f"Downloading Gothic architecture images to: {pic_dir}")
    arch_crawler = GoogleImageCrawler(storage={'root_dir': str(pic_dir)})
    arch_keywords = [
        "Gothic cathedral exterior architecture",
        "Gothic church pointed arches detail",
        "Notre Dame Paris architecture high res"
    ]
    for i, kw in enumerate(arch_keywords):
        arch_crawler.crawl(keyword=kw, max_num=100, file_idx_offset=i * 100)

    print(f"\nDownloading Gothic film stills to: {film_dir}")
    film_crawler = GoogleImageCrawler(storage={'root_dir': str(film_dir)})
    film_keywords = [
        "Gothic horror movie cinematography",
        "Tim Burton movie aesthetic gothic",
        "Gothic film noir scene"
    ]
    for i, kw in enumerate(film_keywords):
        film_crawler.crawl(keyword=kw, max_num=100, file_idx_offset=i * 100)

    print("\nTask completed.")

gothic_dual_crawler()
```
